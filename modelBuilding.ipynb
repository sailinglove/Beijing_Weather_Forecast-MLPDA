```python
import pandas as pd #数据处理
```

Read data from clean data csv.

```python
file = r'data/clean_data.csv'
df = pd.read_csv(file)
df
```

Create small sample DataFrame for easier testing.

```python
tmp = df[['date', 'mintemp', 'meanpres', 'E_wd', 'some']].head(10)
tmp
```

```python
# 1 day prior
N = 2

# target measurement of mean temperature
feature = 'mintemp'

# total number of rows
rows = tmp.shape[0]

# a list representing Nth prior measurements of feature
# notice that the front of the list needs to be padded with N
# None values to maintain the constistent rows length for each N
nth_prior_measurements = [None]*N + [tmp[feature][i-N] for i in range(N, rows)]

# make a new column name of feature_N and add to DataFrame
col_name = "{}_{}".format(feature, N)  
tmp[col_name] = nth_prior_measurements  
tmp  
```

Function for creating the required feature for the n-th prior day.

```python
def get_nth_prior(df, feature, n): 
    rows = df.shape[0]
    nth_prior_measurements = [None]*n + [df[feature][i-n] for i in range(n, rows)]
    col_name = "{}_{}".format(feature, n)
    df[col_name] = nth_prior_measurements
```

```python
features = [column for column in tmp]
print(features)

for feature in features:
    if feature != 'date':
        for n in range(1, 4):
            get_nth_prior(tmp, feature, n)

tmp
```

```python
features = [column for column in df]
print(features)

for feature in features:
    if feature != 'date':
        for n in range(1, 4):
            get_nth_prior(df, feature, n)

df
```

```python
df.info(verbose=True)
```

Drop rows with null value to ensure the integrity of dataset.

```python
df_nona = df.dropna()
df_nona
```

```python
import re

s = r'._\d'

to_keep = df_nona.columns[df_nona.columns.str.contains(s, regex=True)]
to_remove = [column for column in df_nona.columns if column not in to_keep]
x = df_nona.drop(to_remove, axis=1)
x
```

```python
y = df_nona['meantemp']
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train
```

```python
y_train
```

## Basic Linear Regression

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(x_train, y_train)
```

```python
prediction = regressor.predict(x_test)
prediction
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print("The Explained Variance: %.2f" % regressor.score(x_test, y_test))  
print("The Mean Absolute Error: %.2f degrees celsius" % mean_absolute_error(y_test, prediction))  
print("The Mean Squared Error: %.2f degrees celsius" % mean_squared_error(y_test, prediction))
```

Save the trained model as a file by using `joblib`.

```python
import joblib

joblib.dump(regressor, r'models/basicLinearRegression.model')

joblib.dump(x, r'models/x.dataset')
joblib.dump(y, r'models/y.dataset')
joblib.dump(x_train, r'models/x_train.dataset')
joblib.dump(x_test, r'models/x_test.dataset')
joblib.dump(y_train, r'models/y_train.dataset')
joblib.dump(y_test, r'models/y_test.dataset')
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso_regressor = Lasso()

lasso_regressor.fit(x_train, y_train)

lasso_prediction = lasso_regressor.predict(x_test)
lasso_prediction

print("The Explained Variance: %.2f" % lasso_regressor.score(x_test, y_test))  
print("The Mean Absolute Error: %.2f degrees celsius" % mean_absolute_error(y_test, lasso_prediction))  
print("The Mean Squared Error: %.2f degrees celsius" % mean_squared_error(y_test, lasso_prediction))
```

发现误差更大了

这是因为lasso regression对于alpha的取值非常敏感，默认的 `alpha=1` 非最佳

sklearn内建类 `LassoCV` 使用 cross validation 自动找出给定取值范围内的最优 alpha，并使用该 alpha 构建回归模型

https://blog.csdn.net/qq_45797116/article/details/112676554

```python
from sklearn.linear_model import LassoCV

lassoCV_regressor = LassoCV(eps=0.0001)

lassoCV_regressor.fit(x_train, y_train)

lassoCV_prediction = lassoCV_regressor.predict(x_test)
lassoCV_prediction

print("The Explained Variance: %.2f" % lassoCV_regressor.score(x_test, y_test))  
print("The Mean Absolute Error: %.2f degrees celsius" % mean_absolute_error(y_test, lassoCV_prediction))  
print("The Mean Squared Error: %.2f degrees celsius" % mean_squared_error(y_test, lassoCV_prediction))
```

Trail and Error 发现当 `eps` 取值 $0.0001$ 时模型预测误差最小，但似乎精度过高导致模型内部训练时可能出现精度误差，结果未必准确

```python
joblib.dump(lasso_regressor, r'models/lassoRegression.model')
joblib.dump(lassoCV_regressor, r'models/lassoRegressionCV.model')
```

## 筛选线性相关高的features

The aim of this model is to predict the mean temperature with given historical data. By the nature of the linear regression model, the features used in the predictions have to have strong correlation with the target.

Hence, we need to study the correlation between the features and our target, mean temperature *meantemp*:

```python
# pd.set_option('display.max_columns', None)
```

查看

```python
corr = df_nona.corr(numeric_only=True)[['meantemp']].T  #计算特征与目标(meantemp)之间的相关度
corr
```

```python
#筛选出线性相关度小于0.6的特征准备从数据集中去除
to_drop_corr = df_nona.columns[[True] + [(True if \
    (abs(corr[column]['meantemp']) < 0.6) else False) for column in corr]]
to_drop_corr
```

```python
df_corr = df_nona.drop(to_drop_corr, axis=1)
df_corr
```

```python
to_keep_corr = df_corr.columns[df_corr.columns.str.contains(s, regex=True)]
to_remove_corr = [column for column in df_corr.columns if column not in to_keep_corr]
x_corr = df_corr.drop(to_remove_corr, axis=1)
x_corr
```

```python
y_corr = y
```

```python
corr_path = r'models/correlated/'
joblib.dump(x_corr, corr_path + r'xc.dataset')
joblib.dump(y_corr, corr_path + r'yc.dataset')
```

```python
xc_train, xc_test, yc_train, yc_test = train_test_split(x_corr, y_corr, test_size=0.2, random_state=42)
xc_train
```

```python
joblib.dump(xc_train, corr_path + r'xc_train.dataset')
joblib.dump(yc_train, corr_path + r'yc_train.dataset')
joblib.dump(xc_test, corr_path + r'xc_test.dataset')
joblib.dump(yc_test, corr_path + r'yc_test.dataset')
```

### Basic LR with more correlated data
